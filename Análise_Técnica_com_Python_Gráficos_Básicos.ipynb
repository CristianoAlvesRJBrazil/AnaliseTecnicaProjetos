```python
# Importe as bibliotecas no seu script Python
import mplfinance as mpf
import yfinance as yf
import matplotlib.pyplot as plt

# Baixar dados de exemplo
df = yf.download('PETR4.SA', start='2023-01-01', end='2023-12-31')

# Verificar os dado
df.head()
```

```python
# Plotando Gráficos de Candlestick
mpf.plot(df, type='candle', figratio=(40, 15), title='Gráfico de Preços da PETR4.SA')
```

```python
# Plotando Gráficos de Candlestick com Volume
mpf.plot(df, type='candle', volume=True, figratio=(40, 15), title='Gráfico de Preços da PETR4.SA')
```

```python
# Plotando Gráficos de Candlestick com Volume - Personalizando o Estilo do Gráfico com o style ='charles'
mpf.plot(df, type='candle', style='charles', volume=True, figratio=(40, 15), title='Gráfico de Preço da PETR4.SA', ylabel='Preço', ylabel_lower='Volume')
```

```python
# Função para adicionar linhas de suporte e resistência '''
def add_lines(ax):
    ax.axhline(y=38.5, color='g', linestyle='--')
    ax.axhline(y=32.5, color='r', linestyle='--')

fig, axlist = mpf.plot(df, type='candle',style='charles', volume=True, title='Gráfico de Preço da PETR4.SA', ylabel='Preço', ylabel_lower='Volume', figratio=(40, 15), returnfig=True, figscale=1.2)
add_lines(axlist[0])
plt.show()
```

```python
# Gráfico de Linha
mpf.plot(df, type='line',style='charles', title='Gráfico de Preço da PETR4.SA', ylabel='Preço', ylabel_lower='Volume', figratio=(40, 15))
```

```python
# Gráfico de Renko
mpf.plot(df, type='renko',style='charles', title='Gráfico de Preço da PETR4.SA', ylabel='Preço', ylabel_lower='Volume', figratio=(40, 15))
```

```python
# Ponto e Figura (Point and Figure)
mpf.plot(df, type='pnf',style='charles', title='Gráfico de Preço da PETR4.SA', ylabel='Preço', ylabel_lower='Volume', figratio=(40, 15))
```

```python
# Configurando Layouts e Adicionando Subplots
# Calculando a SMA (Média Móvel Simples)
sma_20 = df['Close'].rolling(window=20).mean()

# Configurando o layout com subplots adicionais
add_plots = [mpf.make_addplot(sma_20, color='r')]

# Plotando com subplots adicionais
mpf.plot(df, type='candle', volume=True, addplot=add_plots, style='charles', title='Gráfico de Preço da PETR4.SA', ylabel='Preço', ylabel_lower='Volume', figratio=(40, 15))
```
